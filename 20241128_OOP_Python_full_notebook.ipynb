The point of creating a class is to define some sort of structure. When we create an instance of the class, the method `__init__()` is automatically called. This is where you might want to define some attributes that all instances need to have for your program to work. Arguments can be passed to `__init__()` to make our life easier. Let's redefine our `Character` class.  
*L'objectif quand on créée une classe est de définir une structure. Quand on créée une instance de la class, la méthode `__init__()` est appelée automatiquement. C'est donc à cet endroit qu'on va vouloir définir des attributs que toutes les instances doivent posséder pour que le programme fonctionne. On peut passer des arguments à `__init__()`, ce qui simplifie la définition des attributs. Réécrivons la classe `Character`.*

```python
class Character:
    
    def __init__(self, name, function, ship, birthplace):
        
        self.name = name
        self.function = function
        self.ship = ship
        self.birthplace = birthplace
```

We redefined the default `__init__()` method to do what we want. The word `self` refers to the instance of the class which is created. It is customary to have `self` as the first argument in all the methods of the class in their definition. It is passed automatically when the method is called, you do not need to give it explicitly as an argument.  
*Nous avons rédéfini la méthode par défaut `__init__()` pour faire ce que l'on voulait. Le mot `self` désigne l'instance de la classe qui est créée. Traditionnellement, on passe `self` comme premier argument à toutes les méthodes de la classe dans leur définition. Il est passé par défaut quand les méthodes sont appelées, pas besoin de le redonner explicitement comme argument par la suite.* 

```python
captain = Character("James T. Kirk", "Captain", "USS Enterprise", "Earth")
doctor = Character("Beverly Crusher", "Chief Medical Officer", "USS Enterprise D", "Luna")
```

```python
print(f"{doctor.name} is the {doctor.function} aboard the {doctor.ship}. This character was born on {doctor.birthplace}.")
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as pim

class Character:
    
    def __init__(self, name, function, ship, birthplace):
        """
        Parameters
        ----------
        name: str
            Name of the character.
        function: str
            Function of the character on the ship, like "Captain", "Science Officer", etc.
        ship: str
            Name of the ship on which the character typical lives.
        birthplace: str
            Planet/moon/whatever of birth of the character.
        """
        
        self.name = name
        self.function = function
        self.ship = ship
        self.birthplace = birthplace

    def __str__(self):
        return f"Instance of class Character. {self.name} is the {self.function} aboard the {self.ship}. This character was born on {self.birthplace}."

    def say_hello(self, picture, quote):
        """
        Parameters
        ----------
        picture: str
            Path to the characters picture.
        quote: str
            Typical quote of the character
        """
        text_quote = f"Hello I'm {self.name}!\n{quote}"
        image = pim.imread(picture) 
        fig, ax = plt.subplots(1, 1, figsize=(8, 6))
        ax.imshow(image)
        ax.axis('off')
        props = {'boxstyle': 'round', 'facecolor': 'wheat', 'pad': 0.75}
        ax.text(0.1, 0.85, text_quote, transform=ax.transAxes, fontsize=14, bbox=props)
        plt.show()
```

## Class methods and attributes / *Méthodes et attributs de classe*

A class method is shared by all instances of the class. There is no need to create an instance of the class to call a class method. A class method is specified using the `@classmethod` decorator. The first argument of a class method is the class itself. By convention, its name is `cls`.  
*Une méthode de classe est partagée par toutes les instances de la classe. Il n'est pas nécessaire de créer une instance de la classe pour appeler une méthode de classe. On définit une méthode de classe avec le décorateur `@classmethod`. Le premier argument d'une méthode de classe est la classe elle-même. Par convention, on l'appelle `cls`.*

```python
import random

class Character:
    
    def __init__(self, name, function, ship, birthplace):
        """
        Parameters
        ----------
        name: str
            Name of the character.
        function: str
            Function of the character on the ship, like "Captain", "Science Officer", etc.
        ship: str
            Name of the ship on which the character typical lives.
        birthplace: str
            Planet/moon/whatever of birth of character.
        """
        
        self.name = name
        self.function = function
        self.ship = ship
        self.birthplace = birthplace

    def __str__(self):
        return f"Instance of class Character. {self.name} is the {self.function} aboard the {self.ship}. This character was born on {self.birthplace}."

    def say_hello(self, picture, quote):
        """
        Parameters
        ----------
        picture: str
            Path to the characters picture.
        quote: str
            Typical quote of the character
        """
        text_quote = f"Hello I'm {self.name}!\n{quote}"
        image = pim.imread(picture) 
        fig, ax = plt.subplots(1, 1, figsize=(8, 6))
        ax.imshow(image)
        ax.axis('off')
        props = {'boxstyle': 'round', 'facecolor': 'wheat', 'pad': 0.75}
        ax.text(0.1, 0.85, text_quote, transform=ax.transAxes, fontsize=14, bbox=props)
        plt.show()

    @classmethod
    def create_random_character(cls):
        with open('characters.csv', 'r') as f:
            content = f.readlines()
        person = random.choice(content).split('\t')
        person = [element.strip() for element in person]
        return cls(*person)
```

```python
print(Character.create_random_character())
```

Similarly, class attributes are variables that you define directly in the class body but outside of any method. These attributes are tied to the class itself rather than to particular instances of that class. For example, we can add a counter for the number of `Characters` created.  
*De façon similaire, les attributs de classe sont des variables qu'on définit directement dans le corps de la classe, hors des méthodes. Ces attribus sont attachés à la classe elle-même plutôt qu'à des instances particulières. Par exemple, on peut ajouter un compteur du nombre de `Character` créés.*

```python
class Character:

    charac_nb = 0
    
    def __init__(self, name, function, ship, birthplace):
        """
        Parameters
        ----------
        name: str
            Name of the character.
        function: str
            Function of the character on the ship, like "Captain", "Science Officer", etc.
        ship: str
            Name of the ship on which the character typical lives.
        birthplace: str
            Planet/moon/whatever of birth of character.
        """
        Character.charac_nb +=1
        
        self.name = name
        self.function = function
        self.ship = ship
        self.birthplace = birthplace

    def __str__(self):
        return f"Instance of class Character. {self.name} is the {self.function} aboard the {self.ship}. This character was born on {self.birthplace}."

    def say_hello(self, picture, quote):
        """
        Parameters
        ----------
        picture: str
            Path to the characters picture.
        quote: str
            Typical quote of the character
        """
        text_quote = f"Hello I'm {self.name}!\n{quote}"
        image = pim.imread(picture) 
        fig, ax = plt.subplots(1, 1, figsize=(8, 6))
        ax.imshow(image)
        ax.axis('off')
        props = {'boxstyle': 'round', 'facecolor': 'wheat', 'pad': 0.75}
        ax.text(0.1, 0.85, text_quote, transform=ax.transAxes, fontsize=14, bbox=props)
        plt.show()

    @classmethod
    def create_random_character(cls):
        with open('characters.csv', 'r') as f:
            content = f.readlines()
        person = random.choice(content).split('\t')
        person = [element.strip() for element in person]
        return cls(*person)
```

```python
print(Character.create_random_character())
print(f"Number of characters created {Character.charac_nb}.")
```

```python
spock = Character('Spock', 'Science Officer', 'USS Enterprise', 'Vulcan')
print(spock)
print(f"Number of characters created {Character.charac_nb}.")
```

# Exercise / *Exercice*

Create a class `Ship`, with an attribute `maximal_warp` which is a float between 0 and 9.99, and another attribute `crew` which is a list of `Character` (who have the correct `ship` attribute, meaning that you need to change it if necessary). Caution, a ship needs a captain!  
*Créer une classe `Ship` avec un attribut `maximal_warp` qui est un float entre 0 et 9.99, et un autre attribut `crew` qui est une liste de `Character` (avec le bon attribut `ship`, donc il faut le changer si nécesaire). Attention, un vaisseau a besoin d'un capitaine!*  

Add a method `beam_to` which takes a list of `Character` and another `Ship` as arguments, and transfers the characters, if they are onboard, to the indicated ship.  
*Ajouter une méthode `beam_to` qui prend comme argument une list de `Character` et un autre `Ship` comme arguments et transfère les `Character`, s'ils sont à bord, sur le vaisseau indiqué.*

```python
class Ship:

    def __init__(self, name, max_warp, crew):
        """
        Parameters
        ----------
        max_warp: float
            Maximum warp speed, float between 0 and 9.99.
        crew: list of Character
            Crew in the ship, a captain needs to be in the list, otherwise a randomly chosen crew member is promoted. 
            Also it cannot be empty, otherwise we just create a random guy and declare they are the captain.
        """
        self.name = name
        self._maximum_warp = self.check_warp_speed(max_warp)

        self.crew = []
        captain_found = False
        for c in crew:
            c.ship = self.name # we assign the crew to the ship
            if "Captain" in c.function:
                captain_found = True
            self.crew.append(c)
                
        if len(self.crew) == 0: # we cannot have an empty crew, let's create a random character
            self.crew.append(Character.create_random_character())
                
        if not captain_found: # we need a captain
            new_captain = random.choice(self.crew)
            new_captain.function = "Captain"

    def __str__(self):
        crew_str = ""
        for c in self.crew:
            crew_str += f"- {c.name}, who is the {c.function} onboard.\n"
        return f"Instance of {type(self)}. The {self.name} has a maximum warp speed of {self.maximum_warp}. Its crew is composed of:\n"+crew_str

    @property
    def maximum_warp(self):
        return self._maximum_warp

    @maximum_warp.setter
    def maximum_warp(self, max_warp):
        self._maximum_warp = self.check_warp_speed(max_warp)

    def check_warp_speed(self, max_warp):
        if max_warp < 0:
            return 0
        elif max_warp > 10:
            print('We do not have the technology to go above warp 9.99!')
            return 9.99
        else: 
            return max_warp

    def add_crew_members(self, crew_members):
        """
        Parameters
        ----------
        crew member: list of Character
            The crew members to add to the ship's crew list.
        """
        self.crew.extend(crew_members)

    def beam_to(self, crew_to_beam_up, ship):
        """
        Parameters
        ----------
        crew_to_beam_up: list of str
            List of names of crew members to beam up.
        ship: Ship
            The ship to which we want to beam the crew.
        """
        crew_names = [c for c in crew_to_beam_up]
        transporter_room = []
        
        for i, c in enumerate(self.crew):
            if c.name in crew_names:
                transporter_room.append((i, c))

        # we need to remove characters from the list in reverse order to avoid messing up the indices
        transporter_room.reverse() 
        print('Energize!')
        
        for c in transporter_room:
            c[1].ship = ship.name
            self.crew.pop(c[0])
        ship.add_crew_members([c[1] for c in transporter_room])

        if len(self.crew) == 0:
            print('There is no one left on the ship! It might be taken over by Romulans!')
```

# Inheritance / Héritage

Inheritance is the process by which one class takes on the attributes and methods of another (or several). Newly formed classes are called child classes, and the classes that you derive child classes from are called parent classes. It allows easy reusability of code and avoiding to write the same code again and again.  
*L'héritage est le processus qui permet à une classe de posséder les attributs et méthodes d'une autre classe (ou plusieurs). Les nouvelles classes ainsi formées sont appelées classes filles, et les classes dont elles sont dérivées sont les classes parentes. Ce système permet une réutilisation facile du code et ainsi d'éviter de réécrire le même code un grand nombre de fois.*

Let's create child classes of ships. We want a class `Shuttle` for small travels, which can only accomodate maximum 2 crew members. We also want a new class `Starship`, for big ships that travel far. These guys require a minimum crew of 5, including a Captain, a First Officer and a Chief Medical Officer. 
*Créons des classes filles de `Ship`. Nous voulons une classe `Shuttle` pour les petits trajets, pour des vaisseaux qui ne peuvent être occupés que par 2 personnes maximum. Nous voulons aussi une nouvelle classe `Starship` pour les gros vaisseaux qui voyagent loin. Ces engins nécessitent un équipage d'au moins 5 personnes, dont un capitaine, un second, et un docteur.* 

```python
class Shuttle(Ship):
    
    def __init__(self, name, max_warp, crew): # this overrides the __init__ function of Ship!
        
        if len(crew) > 2:
            print('This crew is too big! Only the 2 first members will be admitted onboard.')
            crew = crew[:2]
            
        super().__init__(name, max_warp, crew) # call the __init__ of the parent class.
        

    def add_crew_members(self, crew_members):
        """
        Parameters
        ----------
        crew member: list of Character
            The crew members to add to the ship's crew list.
        """
        if len(self.crew) == 2:
            print(f'Already 2 crew members in the shuttle {self.name}. The characters were beamed into outer space!')
        elif len(self.crew) + len(crew_members) > 2:
            n = max(0, 2-len(self.crew))
            self.crew.extend(crew_members[:n])
            print('The total of people to beam up and already there is exceeding 2.\n',
                  f'Only the first {n} characters reached the shuttle {self.name}, the others were beamed into outer space!')
        else:
            self.crew.extend(crew_members)
```

```python
class Starship(Ship):

    def __init__(self, name, max_warp, crew): # this overrides the __init__ function of Ship!

        if len(crew) < 5:
            print('This crew is too small, we will add random characters!')
            while len(crew) < 5:
                crew.append(Character.create_random_character())
                  
        super().__init__(name, max_warp, crew) # call the __init__ of the parent class.

        self.check_and_promote_crew()

    
    def check_and_promote_crew(self):
        """
        Checks if we have a Captain, a First Officer and a Chief Medical Officer onboard, otherwise
        promotes random crew members.
        """
        core_crew_indices = {'Captain': -1, 
                             'First Officer': -1,
                             'Chief Medical Officer': -1}
        
        for i, c in enumerate(self.crew):
            for k in core_crew_indices.keys():
                if c.function == k:
                    core_crew_indices[k] = i
            
        
        crew_indices = [i for i in range(len(self.crew))]
        # We do not want to lose our captain, first officer or doctor in the promotion process!
        for k in core_crew_indices.keys():  
            if core_crew_indices[k] > 0:
                crew_indices.remove(core_crew_indices[k])

        # Let's promote random guys
        for k in core_crew_indices.keys():  
             if core_crew_indices[k] == -1: # we miss this function
                new_off = random.choice(crew_indices)
                print(f"{self.crew[new_off].name}, previously {self.crew[new_off].function} is now the {k}!")
                self.crew[new_off].function  = k
                crew_indices.remove(new_off)

        
    def add_crew_members(self, crew_members):
        """
        Parameters
        ----------
        crew member: list of Character
            The crew members to add to the ship's crew list.
        """
        super().add_crew_members(crew_members)
        self.check_and_promote_crew()

    
    def beam_to(self, crew_to_beam_up, ship):
        """
        Parameters
        ----------
        crew_to_beam_up: list of str
            List of names of crew members to beam up.
        ship: Ship
            The ship to which we want to beam the crew.
        """
        
        if len(self.crew) - len(crew_to_beam_up) < 5:
            print('We cannot beam so many people away, we could not operate the Starship with such a small crew!', 'Transport aborted!')
            return
        else:
            super().beam_to(crew_to_beam_up, ship)
            
```
